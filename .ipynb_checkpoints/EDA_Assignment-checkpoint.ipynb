### Univariate Visualization
* Plot two Charts
* Provide each charts with at least two insights 

### CHART 1

```python
#df['Population estimates'] = df['Population estimates'].str.replace(',', '').astype(float)
nationality_data = df.groupby('Nationality')['Population estimates'].sum()
plt.figure(figsize=(8, 8))
plt.pie(nationality_data, labels=nationality_data.index, autopct='%1.1f%%', startangle=140)
plt.title('Population Distribution by Nationality')
plt.show()
```

#### Write your insghts here

- The percentage of non-Saudis is close to the percentage of Saudis
- The proportion of Saudis represents approximately half of the population of the Kingdom of Saudi Arabia

```python
region_population = df.groupby('Region')['Population estimates'].sum()
plt.figure(figsize=(12, 6))
region_population.plot(kind='bar', color='purple')
plt.title('Distribution of Regions')
plt.xlabel('Region')
plt.ylabel('Total Population')
plt.xticks(rotation=90)

plt.show()
```

#### Write your insghts here

- The highest population in the Kingdom is in the capital, Riyadh
- The main governorates of Riyadh, Mecca, and the Eastern Province have the highest population in the Kingdom

### Bivariate Visualization 
* Plot two Charts
* Provide each charts with at least two insights 

```python
population = df.groupby('Year')['Population estimates'].sum()

plt.figure(figsize=(10, 6))
plt.plot(population.index, population.values, marker='o', linestyle='-', color='purple')

plt.title('Total Population by Year')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.grid(True)

plt.show()
```

#### Write your insghts here

- The rise was increasing annually and began to decline in 2017 and then returned to rise in 2020
- The year 2022 recorded the highest population rate since 2010

```python
counts = df.groupby(['Region', 'Nationality'])['Population estimates'].sum().unstack().fillna(0)

counts.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Population by Nationality in Each Region')
plt.xlabel('Region')
plt.ylabel('Population')
plt.legend(title='Nationality')
plt.show()
```

#### Write your insghts here

- This figure compares the number of Saudis to non-Saudis in all regions
- It appears that the largest concentration of non-Saudis is in the three main regions(Riyadh, Mecca, and the Eastern Provinc), where their number is close to the number of Saudis

### Multivariate Visualization 
* Plot two Charts
* Provide each charts with at least two insights 

```python
gro_sum = df.groupby(['Region', 'Year'])['Population estimates'].sum().reset_index()
fig = gro_sum['Population estimates'] / 1000

plt.figure(figsize=(20, 20))
plt.scatter(gro_sum['Year'], gro_sum['Region'], s=fig, alpha=0.5)
plt.title('Scatter Plot of Region vs Year (Population-based)')
plt.xlabel('Year')
plt.ylabel('Region')
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

#### Write your insghts here

- The population in some areas has not changed significantly
- The increase is clear in the Makkah and Riyadh Regions

```python
gro_sum = df.groupby(['Region', 'Year'])['Population estimates'].sum().reset_index()
fig = gro_sum.pivot(index='Region', columns='Year', values='Population estimates')

fig.plot(kind='bar', figsize=(12, 8))

plt.title('Population Distribution by Region and Year')
plt.xlabel('Region')
plt.ylabel('Population')
plt.xticks(rotation=45)

plt.show()
```

#### Write your insghts here

- The difference between the population in 2010 and 2022 is clear
- The highest increase in population is in the Riyadh region
